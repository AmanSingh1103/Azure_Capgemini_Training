```python
# Step 1: Install latest SDK
!pip install --upgrade openai

# Step 2: Imports
from google.colab import userdata
from openai import OpenAI

# Step 3: Load API key (make sure key name is correct)
api_key = userdata.get("aaif")   # вљ пёЏ ensure this matches your saved key name

# Step 4: Azure AI Foundry endpoint
endpoint = "https://email-assistant-resource.services.ai.azure.com/api/projects/email-assistant"

# Step 5: Agent details
agent_name = "AI-InterviewSimulator"
version = "1"

# Step 6: Create client
client = OpenAI(
    api_key=api_key,
    base_url=f"{endpoint}/openai/v1"
)

# Step 7: Continuous chat loop (better than single input)
print("рџЋ¤ AI Interview Simulator Started (type 'exit' to stop)\n")

while True:
    user_input = input("рџ‘¤ Your Answer: ")

    if user_input.lower() == "exit":
        print("Interview Ended.")
        break

    # Step 8: Send request to agent
    response = client.responses.create(
        input=user_input,
        extra_body={
            "agent_reference": {
                "name": agent_name,
                "version": version,
                "type": "agent_reference"
            }
        }
    )

    # Step 9: Print response
    print("\nрџ¤– Interviewer:\n")
    print(response.output_text)
    print("\n" + "-"*50 + "\n")
```
